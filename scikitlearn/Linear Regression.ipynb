```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
```

```python
%matplotlib notebook
```

```python
data = pd.read_csv('data.csv')
```

```python
data.plot(kind='scatter', x='studytime', y='score')
```

```python
# Test train split for supervised tranning
X_train, X_test, y_train, y_test = train_test_split(data.studytime, data.score)
```

```python
plt.scatter(X_train, y_train, label='Tranning data', color='r', alpha=.7)
plt.scatter(X_test, y_test, label='Testing data', color='g', alpha=.7)
plt.legend()
plt.title('Test Train split')
plt.show()
```
